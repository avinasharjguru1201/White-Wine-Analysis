# Data description

The data is in csv format.In computing, a comma-separated values (CSV) file stores tabular data (numbers and text) in plain text. Each line of the file is a data record. Each record consists of one or more fields, separated by commas.
White wine is mainly from "white" grapes, which are green or yellow in colour, such as the Chardonnay, Sauvignon, and Riesling. Some white wine is also made from grapes with coloured skin, provided that the obtained wort is not stained.
Data are collected on 12 different properties of the wines one of which is Quality, based on sensory data, and the rest are on chemical properties of the wines including density, acidity, alcohol content etc. All chemical properties of wines are continuous variables. Quality is an ordinal variable with possible ranking from 1 (worst) to 10 (best). Each variety of wine is tasted by three independent tasters and the final rank assigned is the median rank given by the tasters.

# Application of Classification models


```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,log_loss
```

## Seperating the label column form the train data for training and testing

```python
label = df['quality']

train = df.drop(columns = ['quality'])
```

## Splitting the data for training and testing

```python
from sklearn.cross_validation import train_test_split

x_train , x_test , y_train , y_test = train_test_split(train , label , test_size = 0.30)
```

## Decision Tree Classifier

### Preparing the model and importing necessary packages

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

### Fitting the model

```python
clf.fit(x_train, y_train)
```

### We find the accuracy score

```python
confidence = clf.score(x_test, y_test)
print("\nAccuracy:\n")
print(confidence)
```

```python
pro = clf.predict_proba(x_test)
```

```python
y_pred = clf.predict(x_test)
```

### Log Loss

```python
log_loss(y_test,pro)
```

### Mean Absolute Error

```python
mean_absolute_error(y_test, y_pred)
```

### Mean Squared Error

```python
mean_squared_error(y_test, y_pred)
```

## Gradient Boosting Classifier

### Preparing the model and importing necessary packages

```python
from sklearn import model_selection as ms
from sklearn.ensemble import GradientBoostingClassifier
```

```python
grad = GradientBoostingClassifier()
```

### Fitting the model

```python
grad.fit(x_train,y_train)
```

### We find the accuracy score

```python
acc = grad.score(x_test,y_test)

print("\nAccuracy:\n")
print(acc)
```

```python
pro = grad.predict_proba(x_test)
```

```python
y_pred = grad.predict(x_test)
```

### Log Loss

```python
log_loss(y_test,pro)
```

### Mean Absolute Error

```python
mean_absolute_error(y_test, y_pred)
```

### Mean Squared Error

```python
mean_squared_error(y_test, y_pred)
```

## Bagging Classifier

### Preparing the model and importing necessary packages

```python
from sklearn.ensemble import BaggingClassifier
from sklearn import model_selection as ms
```

```python
cart = DecisionTreeClassifier()

model = BaggingClassifier(base_estimator = cart,n_estimators = 50,random_state = 10)
```

### Fitting the model

```python
model.fit(x_train,y_train)
```

### We find the accuracy score

```python
acc1 = model.score(x_test,y_test)

print("\nAccuracy:\n")
print(acc1)
```

```python
pro = model.predict_proba(x_test)
```

```python
y_pred = model.predict(x_test)
```

### Log Loss

```python
log_loss(y_test,pro)
```

### Mean Absolute Error

```python
mean_absolute_error(y_test, y_pred)
```

### Mean Squared Error

```python
mean_squared_error(y_test, y_pred)
```

## Model Comparision

```python
model=['DecisionTree','GradientBoostingClassifer','BaggingClassifier',]

score = [confidence,acc,acc1]

compare=pd.DataFrame({'Model':model,'Score':score},index=[i for i in range(1,4)])
```
